```python
# ==============================================================================
# FUNCTION: register_user
# ==============================================================================
def register_user():
    entered_name = ""

    # Keep asking until the user provides a non-empty name
    while entered_name.strip() == "":
        entered_name = input("Please enter your name to begin: ").strip()
        if entered_name == "":
            print(" Name cannot be empty. Please try again.")

    formatted_name = entered_name.title()
    session_memory["username"] = formatted_name

    print()
    print(f" Welcome aboard, {formatted_name}!")
    print("Type 'help' to see available commands, or 'exit' to quit.\n")

    return formatted_name
```

```python
# ==============================================================================
# FUNCTION: save_chat_to_history
# ==============================================================================
def save_chat_to_history(user_message, bot_message):
    
    try:
        with open(CHAT_HISTORY_FILE, "a", encoding="utf-8") as history_file:
            timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            history_file.write(f"[{timestamp}]\n")
            history_file.write(f"User: {user_message}\n")
            history_file.write(f"Bot : {bot_message}\n")
            history_file.write(SUB_SEPARATOR_LINE + "\n")
    except OSError as file_error:
        # If the file cannot be written, notify the user without crashing
        print(f" Warning: Could not save chat history. ({file_error})")

```

```python
# ==============================================================================
# FUNCTION: display_chat_history
# ==============================================================================
def display_chat_history():
    print(SUB_SEPARATOR_LINE)
    print("CHAT HISTORY")
    print(SUB_SEPARATOR_LINE)

    if not os.path.exists(CHAT_HISTORY_FILE):
        print("No chat history found yet. Start chatting to build one!")
    else:
        try:
            with open(CHAT_HISTORY_FILE, "r", encoding="utf-8") as history_file:
                print(history_file.read())
        except OSError as file_error:
            print(f" Could not read chat history. ({file_error})")

    print(SUB_SEPARATOR_LINE)
```

```python
# ==============================================================================
# FUNCTION: display_statistics
# ==============================================================================
def display_statistics():
    elapsed_seconds = time.time() - session_memory["session_start_time"]
    minutes, seconds = divmod(int(elapsed_seconds), 60)

    print(SUB_SEPARATOR_LINE)
    print("SESSION STATISTICS")
    print(SUB_SEPARATOR_LINE)
    print(f"{'Total Questions Asked:':<28}{session_memory['total_questions']}")
    print(f"{'Successful Responses:':<28}{session_memory['successful_responses']}")
    print(f"{'Unknown Commands:':<28}{session_memory['unknown_commands']}")
    print(f"{'Calculator Uses:':<28}{session_memory['calculator_usage']}")
    print(f"{'Session Duration:':<28}{minutes}m {seconds}s")
    print(SUB_SEPARATOR_LINE)

```

```python
# ==============================================================================
# FUNCTION: run_calculator
# ==============================================================================
def run_calculator(operation_name):
    session_memory["calculator_usage"] += 1
    try:
        # Square root only needs a single number
        if operation_name == "sqrt":
            number_value = float(input("Enter a number: "))
            if number_value < 0:
                print("Bot:  Cannot calculate the square root of a negative number.")
            else:
                calculation_result = math.sqrt(number_value)
                print(f"Bot:  Result = {calculation_result}")
            return

        # All other operations need two numbers
        first_number = float(input("Enter first number: "))
        second_number = float(input("Enter second number: "))

        if operation_name == "add":
            calculation_result = first_number + second_number
        elif operation_name == "subtract":
            calculation_result = first_number - second_number
        elif operation_name == "multiply":
            calculation_result = first_number * second_number
        elif operation_name == "divide":
            if second_number == 0:
                print("Bot:  Cannot divide by zero.")
                return
            calculation_result = first_number / second_number
        elif operation_name == "modulus":
            if second_number == 0:
                print("Bot:  Cannot perform modulus by zero.")
                return
            calculation_result = first_number % second_number
        elif operation_name == "exponent":
            calculation_result = first_number ** second_number
        else:
            print("Bot:  Unknown calculator operation.")
            return

        print(f"Bot:  Result = {calculation_result}")

    except ValueError:
        # Raised when input() cannot be converted to float
        print("Bot:  Invalid input. Please enter numeric values only.")
    except OverflowError:
        # Raised for extremely large exponent calculations
        print("Bot:  The result is too large to calculate.")

```

```python
# ==============================================================================
# FUNCTION: display_calculator_menu
# ==============================================================================
def display_calculator_menu():
    print(SUB_SEPARATOR_LINE)
    print(" CALCULATOR MODULE")
    print(SUB_SEPARATOR_LINE)
    print("Available operations: add, subtract, multiply, divide, modulus, exponent, sqrt")

    chosen_operation = input("Choose an operation: ").lower().strip()

    if chosen_operation in CALCULATOR_OPERATIONS:
        run_calculator(chosen_operation)
    else:
        print("Bot:  Invalid operation. Type 'calculator' again to retry.")
```

```python
# ==============================================================================
# FUNCTION: search_knowledge_base
# ==============================================================================
def search_knowledge_base(knowledge_dictionary, user_input):
    for keyword, response_text in knowledge_dictionary.items():
        if keyword in user_input:
            return response_text
    return None
```

```python
# ==============================================================================
# FUNCTION: generate_bot_response
# ==============================================================================
def generate_bot_response(raw_user_input):
    user_input = raw_user_input.lower().strip()
    username = session_memory["username"]

    # --------------------------------------------------------------------
    # RULE GROUP 1: Greetings
    # --------------------------------------------------------------------
    if user_input in ["hi", "hello", "hey", "salam", "assalamualaikum"]:
        bot_reply = f"Hello {username}! How can I help you today?"

    # --------------------------------------------------------------------
    # RULE GROUP 2: Personal / Identity Questions
    # --------------------------------------------------------------------
    elif "your name" in user_input:
        bot_reply = "My name is DecodeBot. I am a Rule-Based AI Chatbot."

    elif "who created you" in user_input or "who made you" in user_input:
        bot_reply = "I was created by an AI Engineering Intern using pure Python as part of the DecodeLabs internship."

    elif "how are you" in user_input:
        bot_reply = "I am functioning perfectly. Thanks for asking!"

    elif "thank" in user_input:
        bot_reply = "You're welcome! Let me know if there's anything else you'd like to learn."
        # --------------------------------------------------------------------
    # RULE GROUP 3: AI / ML / DL Knowledge Base
    # --------------------------------------------------------------------
    elif search_knowledge_base(KNOWLEDGE_BASE_AI, user_input) is not None:
        bot_reply = search_knowledge_base(KNOWLEDGE_BASE_AI, user_input)

    # --------------------------------------------------------------------
    # RULE GROUP 4: Python Knowledge Base
    # --------------------------------------------------------------------
    elif search_knowledge_base(KNOWLEDGE_BASE_PYTHON, user_input) is not None:
        bot_reply = search_knowledge_base(KNOWLEDGE_BASE_PYTHON, user_input)

    # --------------------------------------------------------------------
    # RULE GROUP 5: Developer Tools Knowledge Base
    # --------------------------------------------------------------------
    elif search_knowledge_base(KNOWLEDGE_BASE_TOOLS, user_input) is not None:
        bot_reply = search_knowledge_base(KNOWLEDGE_BASE_TOOLS, user_input)

    # --------------------------------------------------------------------
    # RULE GROUP 6: Internship / Career Knowledge Base
    # --------------------------------------------------------------------
    elif search_knowledge_base(KNOWLEDGE_BASE_CAREER, user_input) is not None:
        bot_reply = search_knowledge_base(KNOWLEDGE_BASE_CAREER, user_input)

    # --------------------------------------------------------------------
    # RULE GROUP 7: Date and Time
    # --------------------------------------------------------------------
    elif "date" in user_input:
        today_date = datetime.date.today()
        bot_reply = f"Today's date is {today_date.strftime('%A, %d %B %Y')}."

    elif "time" in user_input:
        current_time = datetime.datetime.now().strftime("%H:%M:%S")
        bot_reply = f"The current time is {current_time}."

    # --------------------------------------------------------------------
    # RULE GROUP 8: Motivation and Quotes
    # --------------------------------------------------------------------
    elif "motivation" in user_input or "quote" in user_input:
        # Rotate through quotes based on how many questions have been asked
        quote_index = session_memory["total_questions"] % len(MOTIVATION_QUOTES)
        bot_reply = MOTIVATION_QUOTES[quote_index]

    # --------------------------------------------------------------------
    # RULE GROUP 9: Unknown Input
    # --------------------------------------------------------------------
    else:
        bot_reply = "Sorry, I don't understand that yet. Type 'help' to see the available commands."
        session_memory["unknown_commands"] += 1
        return bot_reply

    # If we reach this point, the bot successfully matched a rule
    session_memory["successful_responses"] += 1
    return bot_reply
```

```python
# ==============================================================================
# FUNCTION: display_exit_summary
# ==============================================================================
def display_exit_summary():
    elapsed_seconds = time.time() - session_memory["session_start_time"]
    minutes, seconds = divmod(int(elapsed_seconds), 60)
    username = session_memory["username"]

    print()
    print(SEPARATOR_LINE)
    print(f" Thank you for chatting with DecodeBot, {username}!")
    print(SEPARATOR_LINE)
    print(" SESSION SUMMARY")
    print(f"{'  Questions Asked:':<28}{session_memory['total_questions']}")
    print(f"{'  Successful Responses:':<28}{session_memory['successful_responses']}")
    print(f"{'  Unknown Commands:':<28}{session_memory['unknown_commands']}")
    print(f"{'  Calculator Uses:':<28}{session_memory['calculator_usage']}")
    print(f"{'  Session Duration:':<28}{minutes}m {seconds}s")
    print(" Chat history saved successfully to 'chat_history.txt'.")
    print(SEPARATOR_LINE)
    print(" Goodbye! Have a great day ahead.")
    print(SEPARATOR_LINE)

```

```python
# ==============================================================================
# FUNCTION: run_chat_engine
# ==============================================================================
def run_chat_engine():
    username = session_memory["username"]

    while True:
        user_input = input(f"{username}: ").strip()

        # ------------------------------------------------------------
        # Handle empty input without wasting a "question"
        # ------------------------------------------------------------
        if user_input == "":
            print("Bot: Please type something, or type 'help' for options.")
            continue

        lowered_input = user_input.lower()

        # ------------------------------------------------------------
        # Exit Commands
        # ------------------------------------------------------------
        if lowered_input in EXIT_COMMANDS:
            display_exit_summary()
            break

        # ------------------------------------------------------------
        # Calculator Module
        # ------------------------------------------------------------
        elif lowered_input == "calculator":
            display_calculator_menu()
            continue

        elif lowered_input in CALCULATOR_OPERATIONS:
            session_memory["calculator_usage"] += 0  # counted inside run_calculator
            run_calculator(lowered_input)
            continue
        # ------------------------------------------------------------
        # Chat History
        # ------------------------------------------------------------
        elif lowered_input == "history":
            display_chat_history()
            continue

        # ------------------------------------------------------------
        # Statistics
        # ------------------------------------------------------------
        elif lowered_input == "stats":
            display_statistics()
            continue

        # ------------------------------------------------------------
        # Help Menu
        # ------------------------------------------------------------
        elif lowered_input == "help":
            display_help_menu()
            continue

        # ------------------------------------------------------------
        # General Rule-Based Response Engine
        # ------------------------------------------------------------
        else:
            bot_response = generate_bot_response(user_input)
            print("Bot:", bot_response)

            # Update conversation memory
            session_memory["total_questions"] += 1
            session_memory["previous_questions"].append(user_input)

            # Persist this exchange to the chat history file
            save_chat_to_history(user_input, bot_response)
```

```python
# ==============================================================================
# FUNCTION: main
# ==============================================================================
def main():
    session_memory["session_start_time"] = time.time()

    display_welcome_screen()
    register_user()
    run_chat_engine()
```

```python
# ==============================================================================
# PROGRAM ENTRY POINT
# ==============================================================================
if __name__ == "__main__":
    main()
```
